```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('omicron.csv')
data.head()
```

```python
data.info()
```

```python
data['Date'].unique()
```

```python
begin_date = '9/20/2021'
data[data['Date']==begin_date]
```

```python
latest_date = '1/5/2022'
data[data['Date']==latest_date]
```

```python
negara1 = data[data['Country'] == 'Indonesia']
negara1
```

```python
negaramalay = data[data['Country'] == 'Malaysia']
negaramalay
```

```python
negaraind = negara1[negara1['Percentage of Omicron']>0]
negaraind
```

```python
negaraind['Date'] = pd.to_datetime(negaraind['Date'])
negaramalay['Date'] = pd.to_datetime(negaramalay['Date'])
negaraind.head()
```

```python
negaraind['BaseDate'] = pd.to_datetime('2021-12-13')
negaramalay['BaseDate'] = pd.to_datetime('2021-09-20')
negaraind.head()
```

```python
negaramalay.head()
```

```python
negaraind['HariKe'] = negaraind['Date'] - negaraind['BaseDate']
negaramalay['HariKe'] = negaramalay['Date'] - negaramalay['BaseDate']
negaraind.head()
```

```python
negaraind.HariKe = negaraind.HariKe/pd.Timedelta(1, unit='d')
```

```python
negaraind.head()
```

```python
negaramalay.HariKe = negaramalay.HariKe/pd.Timedelta(1, unit='d')
negaramalay.head()
```

```python
plt.plot(negaraind.HariKe, negaraind['Percentage of Omicron'])
plt.plot(negaramalay.HariKe, negaramalay['Percentage of Omicron'])
plt.xlabel('Hari ke-')
plt.ylabel('Jumlah Kasus')
plt.legend(['Indonesia', 'Malaysia'])
plt.title('Jumlah perbandingan kasus')
plt.show()
```
